```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.axes._axes as axes
sns.set()
```

```python
bike_data = pd.read_csv('bikes.csv')
bike_data.head()
```

```python
bike_data.workingday.value_counts()
```

```python
bike_data['season'] = bike_data['season'].map({1:'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
```

```python
bike_data['season'].value_counts()
```

```python
bike_data['yr'] = bike_data['yr'].map({0:2011, 1: 2012})
```

```python
bike_data.yr.value_counts()
```

```python
bike_data['workingday'] = bike_data['workingday'].map({1:'workday', 0:'holiday'})
```

```python
bike_data['weathersit'] = bike_data['weathersit'].map({1:'clear',
                                                       2:'misty-cloudy',
                                                       3:'light-rain-snow',
                                                       4:'heavy-rain-snow'})
```

```python
bike_data['weathersit'].value_counts()
```

```python
bike_data.head()


## Univariate
```

```python
sns.distplot(bike_data['casual'])
plt.title("Histogram of casual rides")
```

```python
sns.distplot(bike_data['registered'])
plt.title("Histogram of registered rides")
```

```python

## Bivariate
```

```python
sns.boxplot(data=bike_data, y='casual', x='season')
plt.title("Number of Casual Rides Vs Season")
```

```python
sns.boxplot(data=bike_data, y='registered', x='season')
plt.title("Number of Registered Rides Vs Season")
```

```python
sns.boxplot(data=bike_data, y='casual', x='yr')
plt.title("Number of Casual Rides Vs Year")
```

```python
sns.boxplot(data=bike_data, y='registered', x='yr')
plt.title("Number of Registered Rides Vs Year")
```

```python
sns.boxplot(data=bike_data, y='casual', x='workingday')
plt.title("Number of Casual Rides Vs WorkingDay")
```

```python
sns.boxplot(data=bike_data, y='registered', x='workingday')
plt.title("Number of Registered Rides Vs WorkingDay")
```

```python
sns.boxplot(data=bike_data, y='casual', x='weathersit')
plt.title("Number of Casual Rides Vs Weather")
```

```python
sns.boxplot(data=bike_data, y='registered', x='weathersit')
plt.title("Number of Registered Rides Vs Weather")
```

```python
sns.regplot(data=bike_data, x='temp', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})
plt.title("Number of Casual Rides Vs Temperature")
```

```python
sns.regplot(data=bike_data, x='temp', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})
plt.title("Number of Registered Rides Vs Temperature")
```

```python
sns.regplot(data=bike_data, x='hum', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})

plt.title("Number of Casual Rides Vs Humidity")
```

```python
sns.regplot(data=bike_data, x='hum', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})
plt.title("Number of Registered Rides Vs Humidity")
```

```python
sns.regplot(data=bike_data, x='windspeed', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})
```

```python
sns.regplot(data=bike_data, x='windspeed', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3})
```

```python
bike_data.shape
```

```python
f, ax = plt.subplots(1,1)
sns.regplot(data=bike_data, x='temp', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, ax=ax)
sns.regplot(data=bike_data, x='temp', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, label='registered', ax=ax)
ax.set_ylabel("Ride Frequency")
ax.set_title("Variation of Ride Frequency Over Temperature")
ax.legend(["casual", "registered"])
```

```python
f, ax = plt.subplots(1,1)
sns.regplot(data=bike_data, x='hum', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, ax=ax)
sns.regplot(data=bike_data, x='hum', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, label='registered', ax=ax)
ax.set_ylabel("Ride Frequency")
ax.set_title("Variation of Ride Frequency Over Humidity")
ax.legend(["casual", "registered"])
```

```python
f, ax = plt.subplots(1,1)
sns.regplot(data=bike_data, x='windspeed', y='casual', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, ax=ax)
sns.regplot(data=bike_data, x='windspeed', y='registered', y_jitter=0.3, 
                scatter_kws={'alpha': 0.3}, label='registered', ax=ax)
ax.set_ylabel("Ride Frequency")
ax.set_title("Variation of Ride Frequency Over Windspeed")
ax.legend(["casual", "registered"])
```
