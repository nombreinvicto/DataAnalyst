# Bay Wheel's trip data exploration
## Preliminary Wrangling

This document explores a dataset containing bike trip information 
of trips made during 2017 via the Bay Wheels public bicycle 
sharing system in the San Francisco Bay Area. The data has been collected
from the Ford GoBike System Data.

```python
# import all packages and set plots to be embedded inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes._axes as axes
import seaborn as sns
%matplotlib inline
```

```python
# first load the dataset and look at the first few rows
data = pd.read_csv('bikedata.csv')
data.head()
```

```python
# lets explore the datatypes and the number of records in the dataset
data.info()
```

It can be observed that there are some null data present for the 
columns `member_birth_year` and `member_gender`. We will 
programmatically remove these null records.

```python
# drop the null rows
data.dropna(inplace=True)

# finally check the data table properties
data.info()
```

The dataframe info shows that now there are no null records. 
However, a few more steps of data wrangling and cleaning is left. 
They are as follows:

* start and end time data types need to be converted to datetime 
objects.
* member birth years need to be converted to integer data type.

```python
data['start_time'] = pd.to_datetime(data['start_time'])
data['end_time'] = pd.to_datetime(data['end_time'])
data['member_birth_year'] = data['member_birth_year'].astype('int64')
```

```python
# now look at the final cleaned dataset
data.reset_index(inplace=True, drop=True)
data.info()

# now save the cleaned dataset as a csv file
data.to_csv('bikedata_clean.csv', index=False)
```

## Univariate Exploration

### Q1. Let us first look at a univariate distribution of the trip duration in minutes and try to find out what is the most common trip duration.

```python
# lets first plot a cursory histogram for the distribution of the trip duration and also a boxplot to show the position of
# the median within the distribution

f, ax = plt.subplots(1, 2, figsize=(16,5)) # type: _,axes.Axes 
sns.distplot((data['duration_sec']/60), kde=False, hist_kws={'alpha': 1}, ax=ax[0])
ax[0].set_xlabel('duration_minutes');

sns.boxplot(y=(data['duration_sec']/60), ax=ax[1], orient='h')
ax[1].set_xlabel('duration_minutes');
ax[1].set_xlim(0, 40);
```

From the histogram plot on the left, it is evident that the distribution is highly right skewed and a majority of the trip duration is much less than 100 minutes with some spanning across larger values. Let us look at a box-plot of the trip durations in minutes (on the right) to see if the large values are infact outliers. From a zoomed in version of the boxplot, it shows that most of the larger values of trip durations fall out as outliers. Consequently, we will perform a scaling transformation to the x axis so that we can also bring those points into consideration.

```python
# convert durations to minute for convenience and find out limits of the log transformed values
# so that we can create bins for the log transformed histogram later

duration_min = data['duration_sec']/60
duration_min.describe()

log_duration_min = np.log10(duration_min)
log_duration_min.describe()
```

```python
# lets plot the distribution for the trip duration on a log transformed scale

bins = 10**np.arange(0, 3.6, 0.1)
sns.distplot(bins=bins, a=duration_min, kde=False, hist_kws={'alpha':1})
plt.xscale('log')
plt.xlabel('duration_min (min)')
plt.ylabel('Trips')
plt.title('Distribution of trip durations');
```

#### 1. From the distribution of the trip durations, it is evident that most of the trips are around 10 minutes. This is in exact concordance to the boxplot we saw earlier. It would thus be advisable for the company to bolster advertising for those people who make trips around the most frequent duration of 10 minutes. May be the company can expand its network by appealing to more people who frequent the starting and ending stations whose distances fall within that average 10 minutes of trip duration.


### Q2. Now, lets look at how trip frquencies vary across different months.

```python
#lets make a copy of the dataframe
data2 = data.copy()

# create new columns for start day and start month, start hour and rider age
start_month_series = data2.start_time.dt.month_name()
start_day_series = data2.start_time.dt.weekday_name
start_hour_series = data2.start_time.dt.hour
rider_age_series = 2017 - data2.member_birth_year


data2.insert(2, 'start_month',start_month_series)
data2.insert(3, 'start_day',start_day_series)
data2.insert(4, 'start_hour',start_hour_series)
data2.insert(17, 'rider_age',rider_age_series)


# preview the dataframe
data2.head(1)
```

```python
# now lets look at how many different months are there in the dataframe
trip_per_month = data2.start_month.value_counts()
trip_per_month
```

Well, it looks like the data obtained only shows trip durations from the second half of the year 2017 with months starting from June and going upto December. For that second half of the year, the trip frequency distribution is as follows.

```python
# now plot a barchart to show trip freeuncy over different start months

order = ['June', 'July', 'August', 'September', 'October', 'November', 'December']
color = sns.color_palette()[0]
sns.barplot(x=trip_per_month.index, y=trip_per_month.values, color=color, order=order)
plt.xticks(rotation=20);
locs, labels = plt.xticks()
plt.title('Trip frequency distribution over different start months')
plt.ylabel('Trips')
plt.xlabel('month')
plt.gca().set_yticks([])

for loc, label in zip(locs, labels):
    # get the text property for the label to get the correct count
    count = trip_per_month[label.get_text()]
    # print the annotation just below the top of the bar
    plt.text(loc, count+300, str(count), ha = 'center', color = '#FF4500', fontdict={'weight': 'bold'})
```

#### 2. From the barchart, it is evident that October and surrounding months show the highest frequency of bike share trips with the lowest occurances during the months of June (or July). Probably during the summer months the number of people using bike sharing platforms dwindles. It will thus be advisable for the company to target customers on or near the month of October for the second half of the year.

## Bivariate Exploration

```python
# lets first create a new dataframe data3 that will only consists of the features we are interested to explore 
# further and which doesnt contain unwanted features because the features have been replaced with more meaningfull
# engineered features e.g birth year -> rider_age

data3 = data2.drop(labels=['start_time', 'end_time', 'start_station_id', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_latitude', 'end_station_longitude', 'bike_id', 'member_birth_year'], axis=1)
data3['duration_sec'] = duration_min
data3.rename({'duration_sec': 'duration_min'}, inplace=True, axis=1)
data3.head(2)
```

```python
# lets first take a sample of 2000 datapoints from data3 so that our subsequent
# plots are clearer and faster to render

samples = np.random.choice(data3.shape[0], 2000, replace = False)
data3_sampl = data3.loc[samples,:]
```

```python
# next lets look at the quartiles of the quantitative features so that we can remove 
# outliers in our visual analysis. We will discard any observation above 99% quantile

data3.quantile([.1, .25, .5, .75, .99])
```

```python
# Modify the data3_sample dataframe to get rid ouf aoutliers across all numeric features
data3_sampl = data3_sampl[(data3_sampl.duration_min <= 75) & (data3_sampl.start_hour <= 23) & (data3_sampl.rider_age <=66 )]
```

```python
# now lets look at the correlation strength between the quantitative variables
data3.corr('pearson')
```

### Q11. There seems to be a slight negative correlation coefficient between the duration_min and rider_age. Is it true?

```python
# lets plot a scatterplot of duration_min Vs rider_age
sns.regplot(data=data3_sampl, x='rider_age', y='duration_min', y_jitter=0.3, scatter_kws={'alpha': 0.3}, ci=None)
plt.xlabel('rider age (yr)')
plt.ylabel('Trip Duration (min)')
plt.title('Trend of trip duration over rider age (yrs)');
```

#### 11 (i). Yes, it is infact evident from the scatterplot above that as the rider age increases, the trip duration decreases slightly. The reason behind a negative coefficient could be that older riders dont go too far from their starting points, or may be bikes are more chosen by younger, more athletic subscribers. This trend shows that, although there is a negative correlation between the trip duration and rider age, the strength is pretty weak and close to zero and the company should not make any marketing decisions just based on the age of the riders.


```python
plt.hist2d(data=data3_sampl, y='duration_min', x='rider_age')
plt.colorbar()
plt.title('Heat map of duration_min vs rider age')
plt.ylabel('duration_min (min)')
plt.xlabel('rider_age(yr)');
```

#### 11(ii).The heatmap of the trip duration and rider age shows that most riders are around 30 years old and make less than or equal to 10 mintutes trips. This is an important information for the company as it depicts what age group and what average trip duration to target while planning for newer services.

### Q14. How do trip durations vary across user types?

```python
groupby_user = data3_sampl.groupby(by='user_type', as_index=False).mean()
sns.barplot(data=groupby_user, x='user_type', y='duration_min', color=color)
plt.title('Mean trip durations across user types');
locs, labels = plt.xticks()
plt.ylabel('Trip Duration (min)')
plt.xlabel('User Type')
plt.gca().set_yticks([])
plt.ylim(0, 20)
for loc, label in zip(locs, labels):
    # get the text property for the label to get the correct count
    count = groupby_user[groupby_user['user_type'] == label.get_text()]['duration_min'].values[0]
    count = float(f"{count:0.2f}")
    # print the annotation just below the top of the bar
    plt.text(loc, count+0.5, str(count), ha = 'center', color = '#FF4500', fontdict={'weight': 'bold'})
```

#### 14. This is a very interesting finding. Customers on average spend more time in trips as compared to subscribers althought around 90% of the users are actually subscribers. This can be explained because subscribers are regular commuters whereas customers make one off purchases possibly for long distance journeys.
